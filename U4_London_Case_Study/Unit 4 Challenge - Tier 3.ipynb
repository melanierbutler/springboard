```python
#That seems to have removed all null values, but there are still too many Boroughs. It seems possible that some of the
#same boroughs are named differently - checking Zipcode to find out.
clean_properties['Zipcode'].unique().shape
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#I understand the directions state to divide the 1998 price by the 2018 price, which I originally did for my analysis
#However, this seems incredibly counterintuitive - would you not want a larger ratio to reflect a larger increase?
#I am redoing my analysis with the inverse ratio from the instructions. Perhaps we can discuss.
def create_price_ratio(borough):
    prices1998 = borough_properties.loc[(borough_properties['Borough']== borough) & (borough_properties['Year']==1998)]['Price']
    prices2018 = borough_properties.loc[(borough_properties['Borough']== borough) & (borough_properties['Year']==2018)]['Price']
    ratio = prices2018.mean() / prices1998.mean()
    return ratio
    
    
```

```python
borough_ratios = pd.DataFrame({'Borough':boroughs,'Price_Ratio': 0})
for value in borough_ratios.index.values:
    borough_ratios.loc[value,'Price_Ratio'] = create_price_ratio(borough_ratios.loc[value,'Borough'])
sorted_ratios = borough_ratios.sort_values('Price_Ratio',ascending=False).reset_index().drop(columns='index')
sorted_ratios
```

```python
sorted_ratios.iloc[:15].plot(kind='bar', x='Borough', y='Price_Ratio')
plt.ylabel('Price Ratio')
plt.show()
```

```python
sorted_ratios.plot(kind='box')
```

```python
#I'm curious whether there's a correlation between the starting price and the increase in price over time? 
for value in borough_ratios.index.values:
    borough_ratios.loc[value,'1998_Average_Price'] = borough_properties.loc[(borough_properties['Borough']== borough_ratios.loc[value,'Borough']) & (borough_properties['Year']==1998)]['Price'].mean()
    borough_ratios.loc[value,'2018_Average_Price'] = borough_properties.loc[(borough_properties['Borough']== borough_ratios.loc[value,'Borough']) & (borough_properties['Year']==2018)]['Price'].mean()
```

```python
borough_ratios.plot(kind='scatter',x='1998_Average_Price',y='Price_Ratio')
plt.ylabel('Price Ratio')
plt.xlabel('1998 Average Price')
plt.show()
borough_ratios.corr()
```

There does not appear to be a correlation between the starting value of a neighborhood and its increase in value.

```python
borough_ratios.plot(kind='box',y=['1998_Average_Price','2018_Average_Price'], subplots=True)
```
